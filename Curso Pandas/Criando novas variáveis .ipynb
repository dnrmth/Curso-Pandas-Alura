## Criando novas variáveis 

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
```

```python
dados 
```

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados
```

```python
dados['Valor m²'] = (dados['Valor']/dados['Area']).round(2)
```

```python
dados
```

```python
dados['Valor Bruto m²'] = (dados['Valor Bruto']/dados['Area']).round(2)
```

```python
dados
```

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
dados['Tipo de Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados
```

## Excluindo Variáveis 

```python
dados_aux = pd.DataFrame(dados[['Valor Bruto', 'Valor m²', 'Valor Bruto m²', 'Tipo de Agregado']])
```

```python
dados_aux
```

```python
del dados_aux['Valor Bruto']
```

```python
dados_aux
```

```python
dados_aux.pop('Valor Bruto m²')
```

```python
dados.drop([ 'Valor Bruto'], axis = 1, inplace = True)
```
